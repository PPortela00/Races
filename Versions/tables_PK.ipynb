```python
import pandas as pd
```

```python
races = pd.read_excel('146923races.xlsx')
```

```python
racer = races[['name', 'birth_date']]

racer_table = []
for index, row in racer.iterrows():
    if [row['name'], row['birth_date']] not in racer_table:
        racer_table.append([row['name'], row['birth_date']])

racer_table
```

```python
nation = races[['nation']]

nation_table = []
for index, row in nation.iterrows():
    if row['nation'] not in nation_table:
        nation_table.append(row['nation'])

n = 1
for row in nation_table:
    print(f"INSERT INTO nation VALUES ({n}, '{row}');")
    n += 1
```

```python
age_class = races[['age_class']]

ageClass_table = []
for index, row in age_class.iterrows():
    if row['age_class'] not in ageClass_table:
        ageClass_table.append(row['age_class'])

n = 1
for row in ageClass_table:
    print(f"INSERT INTO age_class VALUES ({n}, '{row}');")
    n += 1
```

```python
event = races[['event', 'event_year', 'distance']]

event_table = []
for index, row in event.iterrows():
    if [row['event'], row['event_year'], row['distance']] not in event_table:
        event_table.append([row['event'], row['event_year'], row['distance']])

n = 1
for row in event_table:
    print(f"INSERT INTO event VALUES ({n}, '{row[0]}', {int(row[1])}, {int(row[2])});")
    n += 1
```

```python
sex = races[['sex']]

sex_table = []
for index, row in sex.iterrows():
    if row['sex'] not in sex_table:
        sex_table.append(row['sex'])

for row in sex_table:
    print(f"INSERT INTO sex VALUES ('{row}');")
```
